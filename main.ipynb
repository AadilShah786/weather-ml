```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('weather.csv')
# df = df[::-1]
# Selecting specific columns
columns_list = df.columns.tolist()
print(columns_list)
```

```python
df=df[[columns_list[0],columns_list[1],columns_list[2],columns_list[11],columns_list[12],columns_list[13],columns_list[14],columns_list[15],columns_list[16],columns_list[17],columns_list[18],columns_list[19]]]

print(df.head())
```

```python
import matplotlib.pyplot as plt
columns_list = df.columns.tolist()

num_cols = len(columns_list)
fig, axes = plt.subplots(num_cols, num_cols, figsize=(15, 15))

# Loop through all combinations of columns
for i, col1 in enumerate(columns_list):
    for j, col2 in enumerate(columns_list):
        # Scatter plot only if the columns are different
        if col1 != col2:
            axes[i, j].scatter(df[col1], df[col2])
            axes[i, j].set_xlabel(col1)
            axes[i, j].set_ylabel(col2)
        else:
            # If same column, leave the plot area blank
            axes[i, j].axis('off')

plt.tight_layout()
plt.show()
```

```python
y_values=df['Rainfall']
x_values1=df['Humidity9am']
x_values2=df['Pressure9am']
x_values3=df['Cloud9am']
x_values4=df['Temp9am']

plt.scatter(x_values1, y_values)
plt.scatter((x_values2-1000)*7, y_values)
plt.scatter(x_values3*10, y_values)
plt.scatter(x_values4*5, y_values)


plt.show
```
